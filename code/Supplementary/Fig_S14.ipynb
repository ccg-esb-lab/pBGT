## Multiday end point distributions

```python
%%julia


tend=Int(time_mins[end])
th=4
tpp=Int(60*th)
println(tend)


thisEndDict=1
Ndays=3

function getThisDictMean(thisEndDict0)
    pcncells=0
    ncels=0
    for (pcn,timespcn) in sort(collect(thisEndDict0))
        #println(pcn,"-",timespcn)
        st=pcn*timespcn
        st>0 ? (pcncells+=st;ncels+=timespcn) : 0

    end
    thismean=(pcncells/ncels)
    
end

ncolors=length(sim_ICs)
tcgrad=cgrad(:heat,ncolors )# )


p=plot(size=(600,500))

for (i,Ndays_time_Pcounts) in enumerate(MNdays_time_Pcounts)
    thisICmeans=[]
    thisIC=sim_ICs[i]
    tc=tcgrad[i]
    for day=1:Ndays
        tfinal=tend*(day-1)+tpp
        #println(tfinal)
        thisEndDict=Ndays_time_Pcounts[tfinal]
        tm=getThisDictMean(thisEndDict)
        
        push!(thisICmeans,tm)
    end
    plot!(p,[1,2,3],thisICmeans,label=false,c=tc,lw=4 )
    scatter!(p,[1,2,3],thisICmeans,label=false,zcolor=thisIC,m=(5,:heat,1,Plots.stroke(1, tc),5))
end

plot!(p,cbar=false)
plot!(p,cbar=true,color=tcgrad,clims=extrema(sim_ICs),  colorbar_ticks=[0.,0.5,1.])
plot!(p, cbar=true, colorbar_title="IC")
plot!(p,xticks=(newticks),xlabel="Time (season)",xlim=(1-.2,3+.2))
plot!(p,ylim=(15,19),ylabel="Mean PCN")




figname=pathFigures*"sFig-14B.pdf"
println("Saving figure $figname ..")
savefig(p,figname)

display(p)
```
